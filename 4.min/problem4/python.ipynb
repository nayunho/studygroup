```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

data1 = pd.read_csv('./data/2014년 졸음운전 교통사고.csv',encoding='euc-kr') 
data2 = pd.read_csv('./data/2015년 졸음운전 교통사고.csv',encoding='euc-kr') 
data3 = pd.read_csv('./data/2016년 졸음운전 교통사고.csv',encoding='euc-kr') 
```

```python
#1. 2014년부터 2016년까지의 졸음운전 교통사고 발생 횟수의 총합과 평균 구하기
total = np.sum(data1["사고(건)"])+np.sum(data2["사고(건)"])+np.sum(data3["사고(건)"])
avg=total/3
print(f'총합 = {total}')
print(f'평균 = {avg}')
```

```python
# 2. 각 연도별 졸음운전 교통사고 발생 건수의 추이 시각화하기

cnt=[np.sum(data1["사고(건)"]),np.sum(data2["사고(건)"]),np.sum(data3["사고(건)"])]
plt.plot(["2014년","2015년","2016년"],cnt)
```

```python
# 3. 졸음운전 교통사고 발생 건수가 가장 많은 연도와 해당 연도의 월별 사고 발생 추이 시각화하기
cnt=[np.sum(data1["사고(건)"]),np.sum(data2["사고(건)"]),np.sum(data3["사고(건)"])]

if np.argmax(cnt)==0:
    month=data1["구분"]
    plt.figure(figsize=(10,6))
    plt.bar(month,data1["사고(건)"])
    plt.xticks(rotation=45)
elif np.argmax(cnt)==1:
    month=data2["구분"]
    plt.figure(figsize=(10,6))
    plt.bar(month,data1["사고(건)"])
    plt.xticks(rotation=45)
else:
    month=data3["구분"]
    plt.figure(figsize=(10,6))
    plt.bar(month,data1["사고(건)"])
    plt.xticks(rotation=45)
```
